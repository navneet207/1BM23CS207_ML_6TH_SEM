```python
#importing diabetes dataset
df_diabetes = pd.read_csv('Dataset of Diabetes .csv')
df_diabetes.head()
```

```python
#data Cleaning on Diabetes dataset
#checking missing values
df_diabetes.isnull().sum()
```

```python
#checking any duplicate values
df_diabetes.duplicated()
```

```python
#To display statistical information of all numerical
df_diabetes.describe()
```

```python
#numerical data
numerical_data = df_diabetes.select_dtypes(include=[np.number])
print("Diabetes Numerical Columns:")
print(numerical_data.columns)
```

```python
#categorical data
categorical_data = df_diabetes.select_dtypes(include=['object', 'category'])
print("Diabetes Categorical Columns:")
print(categorical_data.columns)
```
