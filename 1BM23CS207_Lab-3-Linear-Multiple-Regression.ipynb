Implementation – Linear Regression

```python
'''Predict canada's per capita income in year 2020. Use the data file canada_per_capita_income.csv file. 
If required, apply the necessary data processing steps. Using this build a regression model and predict 
the per capita income for canadian citizens in year 2020
'''

import pandas as pd
import math

# Load dataset
data = pd.read_csv("canada_per_capita_income.csv")

# Display first few rows
print(data.head())

# Extract x (year) and y (income)
x = data['year'].values
y = data['per capita income (US$)'].values

n = len(x)

# Calculate required sums
sum_x = sum(x)
sum_y = sum(y)
sum_xy = sum(x * y)
sum_x2 = sum(x * x)

# Calculate slope (m) and intercept (c)
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
c = (sum_y - m * sum_x) / n

print("Slope (m):", m)
print("Intercept (c):", c)

# Regression equation
print("Regression Model: y = mx + c")
print(f"y = {m}x + {c}")

# Predict for year 2020
year = 2020
predicted_income = m * year + c

print("Predicted per capita income for 2020:", predicted_income)
```

```python
''' Predict Salary of the employee. Use the data file salary.csv file.
If required, apply the necessary data processing steps. Using this build a regression model 
and predict the salary of the employee with 12 years of experience.
 '''

import pandas as pd

data = pd.read_csv("salary.csv")

print(data.head())
print(data.columns)
print(data.isnull().sum())

# Remove missing values
data = data.dropna()

x = data['YearsExperience'].astype(float).values
y = data['Salary'].astype(float).values

n = len(x)

sum_x = sum(x)
sum_y = sum(y)
sum_xy = sum(x * y)
sum_x2 = sum(x * x)

m = (n*sum_xy - sum_x*sum_y) / (n*sum_x2 - sum_x**2)
c = (sum_y - m*sum_x) / n
# Predict salary for 12 years of experience
yearsOfExperience = 12
predicted_salary = m * yearsOfExperience + c

print("Predicted salary for 12 years of experience:", predicted_salary)
```

Implementation – Multiple Linear Regression

```python
''' Considering the data file hiring.csv. The file contains hiring statics for a firm such as experience of candidate, 
his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, 
you need to build a Multiple Linear Regression model for HR department that can help them decide salaries for future candidates. 
Using this predict salaries for following candidates, 
2 yr experience, 9 test score, 6 interview score 
12 yr experience, 10 test score, 10 interview score
 '''

import pandas as pd
import numpy as np
pd.set_option('future.no_silent_downcasting', True)
data = pd.read_csv("hiring.csv")
print(data.head())

# Fill missing values
data['experience'] = data['experience'].fillna('zero')
data['test_score'] = data['test_score'].fillna(data['test_score'].mean())

# Convert words to numbers
word_to_num = {
    'zero':0,'one':1,'two':2,'three':3,'four':4,'five':5,
    'six':6,'seven':7,'eight':8,'nine':9,'ten':10,'eleven':11
}

data['experience'] = data['experience'].replace(word_to_num)

# Explicit numeric conversion (prevents warning)
data['experience'] = data['experience'].astype(float)

# Convert all to numeric
data['experience'] = data['experience'].astype(float)
data['test_score'] = data['test_score'].astype(float)
data['interview_score'] = data['interview_score'].astype(float)
data['salary'] = data['salary'].astype(float)

# Prepare X and Y
X = data[['experience','test_score','interview_score']].values
Y = data['salary'].values

n = len(X)

# Add bias column
X_b = np.c_[np.ones((n,1)), X]

# Normal Equation
B = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(Y)

print("Coefficients:", B)

# Predict for (12,10,10)
x_new = np.array([1,12,10,10])
predicted_salary = x_new.dot(B)

print("Predicted Salary:", predicted_salary)
```

```python
''' Considering the data file 1000_companies.csv. The file contains profit statics for a firm such as R&D Spend, Administration, Marketing Spend
and State. Based on these four factors build a Multiple Linear Regression model to predict the profit. Using this predict profit for following,
91694.48  R&D Spend, 515841.3  Administration, 11931.24  Marketing Spend, Florida State
'''

import pandas as pd
import numpy as np

# Load dataset
data = pd.read_csv("1000_Companies.csv")
print("Original Data:")
print(data.head())

# --- Encode categorical variable: True=1, False=0 ---
data_encoded = pd.get_dummies(data, columns=['State'], drop_first=True)

# Ensure boolean columns are converted to 0/1
for col in data_encoded.columns:
    if data_encoded[col].dtype == 'bool':
        data_encoded[col] = data_encoded[col].astype(int)

print("\nEncoded Data:")
print(data_encoded.head())

# Separate features and target
X = data_encoded.drop('Profit', axis=1)
Y = data_encoded['Profit'].values

# --- Feature scaling (manual standardization) ---
X_scaled = (X - X.mean()) / X.std()

# Convert to numpy
X_vals = X_scaled.values
n = len(X_vals)

# Add bias (intercept) column
X_b = np.c_[np.ones((n,1)), X_vals]

# --- Normal Equation for Linear Regression ---
XT = X_b.T
B = np.linalg.inv(XT.dot(X_b)).dot(XT).dot(Y)

print("\nModel coefficients:")
print(B)
```
